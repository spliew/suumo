```python
df1['区']=df1['区'].str.split('川崎市').apply(lambda x: x[1])
```

```python
df = df1.append([df2,df3])
```

```python
df['バス1']=df['駅1'].str.split(' ').apply(lambda x: x[1].replace('バス','').replace('分','') if len(x) > 1 else  None)
df['バス2']=df['駅2'].str.split(' ').apply(lambda x: x[1].replace('バス','').replace('分','') if len(x) > 1 else  None)
df['バス3']=df['駅3'].str.split(' ').apply(lambda x: x[1].replace('バス','').replace('分','') if len(x) > 1 else  None)
```

```python
df['駅1']=df['駅1'].str.split(' ').apply(lambda x: x[0])
df['駅2']=df['駅2'].str.split(' ').apply(lambda x: x[0])
df['駅3']=df['駅3'].str.split(' ').apply(lambda x: x[0])
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.drop([0],axis=0,inplace=True)
```

```python
 names= ['マンション名', '住所', '区', '市町村', '間取り', '間取りDK', '間取りK', '間取りL', '間取りS',
       '築年数', '建物高さ', '階1', '専有面積', '賃料+管理費', '敷/礼/保証金', '路線1', '駅1',
       '徒歩1', '路線2', '駅2', '徒歩2', '路線3', '駅3', '徒歩3', '賃料', '管理費', '敷金',
       '礼金', '保証金', '敷引,償却','バス1','バス2','バス3']
categorial_features = ['区', '間取りDK', '間取りK', '間取りL', '間取りS',
       '路線1', '駅1', '路線2', '駅2', '路線3', '駅3']
```

```python
dropp = ['マンション名', '住所', '間取り','敷金','賃料', '管理費','賃料+管理費','市町村','敷金','敷/礼/保証金',
       '礼金', '保証金', '敷引,償却']
```

```python
tfvocab = [ 
       '築年数', '建物高さ', '階1', '専有面積',
       '徒歩1', '徒歩2', '徒歩3', 'バス1','バス2','バス3',
'区', '間取りDK', '間取りK', '間取りL', '間取りS',
       '路線1', '駅1', '路線2', '駅2', '路線3', '駅3'
]
```

```python
from sklearn import preprocessing
```

```python
lbl = preprocessing.LabelEncoder()
for col in categorial_features:
    df[col] = lbl.fit_transform(df[col].astype(str))
```

```python
# for col in tfvocab:
#     df[col] = lbl.fit_transform(df[col].astype(float))
df[['築年数', '建物高さ', '階1', '専有面積',
       '徒歩1', '徒歩2', '徒歩3','バス1','バス2','バス3']] = df[['築年数', '建物高さ', '階1', '専有面積',
       '徒歩1', '徒歩2', '徒歩3','バス1','バス2','バス3']].apply(pd.to_numeric)
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
```

```python
y = df['賃料+管理費'].apply(pd.to_numeric)
X = df.drop(dropp,axis=1)
```

```python
# lbl.fit_transform(y.astype(float))
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X,y,test_size=0.2) 
```

```python
lgb_param = {
    'task': 'train',
    'boosting_type' : 'gbdt',
    'objective': 'regression',
    'num_boost_round':30000,
    'early_stopping':200,
    'num_leaves': 80,
     'max_depth': 7,
#     'num_class': 8,
    'metric': 'rmse',
#     'max_depth': 20,
#     'num_leaves': 37,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.8,
    'learning_rate':0.001,
    'verbosity': 200    
#     'min_data': 1,
#     'min_data_in_bin': 1
}
```

```python
lgtrain = lgb.Dataset(X_train,y_train,feature_name=tfvocab, categorical_feature= categorial_features)
lgvalid = lgb.Dataset(X_valid,y_valid,feature_name=tfvocab, categorical_feature= categorial_features)
```

```python
lgb_clf = lgb.train(lgb_param,lgtrain, valid_sets=[lgtrain,lgvalid], valid_names=['train','valid'])
```

```python
lgb_clf.predict(X_valid)
```
