```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('suumo_tokyo0.csv', sep = '\t',encoding='utf-16', names= ['Unnamed: 0','mansion', 'address', 'ku', 'ichi', 'madori', 'madoriDK', 'madoriK', 'madoriL', 'madoriS',
       'chikunen', 'tatemono', 'kai1', 'area', 'money', 'siki', 'rosen1', 'eki1',
       'toho1', 'rosen2', 'eki2', 'toho2', 'rosen3', 'eki3', 'toho3', 'chinryo', 'kanri', 'sikin',
       'reikin', 'hoshokin', 'sikibiki'])
df2 = pd.read_csv('suumo_tokyo2.csv', sep = '\t',encoding='utf-16', names= ['Unnamed: 0','mansion', 'address', 'ku', 'ichi', 'madori', 'madoriDK', 'madoriK', 'madoriL', 'madoriS',
       'chikunen', 'tatemono', 'kai1', 'area', 'money', 'siki', 'rosen1', 'eki1',
       'toho1', 'rosen2', 'eki2', 'toho2', 'rosen3', 'eki3', 'toho3', 'chinryo', 'kanri', 'sikin',
       'reikin', 'hoshokin', 'sikibiki'])
df3 = pd.read_csv('suumo_tokyo3.csv', sep = '\t',encoding='utf-16',  names= ['Unnamed: 0','mansion', 'address', 'ku', 'ichi', 'madori', 'madoriDK', 'madoriK', 'madoriL', 'madoriS',
       'chikunen', 'tatemono', 'kai1', 'area', 'money', 'siki', 'rosen1', 'eki1',
       'toho1', 'rosen2', 'eki2', 'toho2', 'rosen3', 'eki3', 'toho3', 'chinryo', 'kanri', 'sikin',
       'reikin', 'hoshokin', 'sikibiki'])
```

```python
df = df1.append([df2,df3])
```

```python
df['eki1']=df['eki1'].str.split(' ').apply(lambda x: x[0])
df['eki2']=df['eki2'].str.split(' ').apply(lambda x: x[0])
df['eki3']=df['eki3'].str.split(' ').apply(lambda x: x[0])
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.drop([0],axis=0,inplace=True)
```

```python
 names= ['Unnamed: 0','mansion', 'address', 'ku', 'ichi', 'madori', 'madoriDK', 'madoriK', 'madoriL', 'madoriS',
       'chikunen', 'tatemono', 'kai1', 'area', 'money', 'siki', 'rosen1', 'eki1',
       'toho1', 'rosen2', 'eki2', 'toho2', 'rosen3', 'eki3', 'toho3', 'chinryo', 'kanri', 'sikin',
       'reikin', 'hoshokin', 'sikibiki']
categorial_features = ['ku', 'madoriDK', 'madoriK', 'madoriL', 'madoriS',
       'rosen1', 'eki1', 'rosen2', 'eki2', 'rosen3', 'eki3']
```

```python
dropp = ['mansion', 'address', 'madori','siki','chinryo', 'kanri','money','ichi']
```

```python
tfvocab = [ 
       'chikunen', 'tatemono', 'kai1', 'area',
       'toho1', 'toho2', 'toho3', 'sikin',
       'reikin', 'hoshokin', 'sikibiki',
'ku', 'madoriDK', 'madoriK', 'madoriL', 'madoriS',
       'rosen1', 'eki1', 'rosen2', 'eki2', 'rosen3', 'eki3'
]
```

```python
from sklearn import preprocessing
```

```python
lbl = preprocessing.LabelEncoder()
for col in categorial_features:
    df[col] = lbl.fit_transform(df[col].astype(str))
```

```python
for col in tfvocab:
    df[col] = lbl.fit_transform(df[col].astype(float))
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
```

```python
y = df['money']
X = df.drop(dropp,axis=1)
```

```python
lbl.fit_transform(y.astype(float))
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X,y,test_size=0.2) 
```

```python
lgb_param = {
    'task': 'train',
    'boosting_type' : 'gbdt',
    'objective': 'regression',
    'num_boost_round':10000,
    'early_stopping':200,
#     'num_leaves': 80
#      'max_depth': 7,
#     'num_class': 8,
    'metric': 'rmse',
#     'max_depth': 20,
#     'num_leaves': 37,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.8,
    'learning_rate':0.003,
    'verbosity': 200    
#     'min_data': 1,
#     'min_data_in_bin': 1
}
```

```python
lgtrain = lgb.Dataset(X_train,y_train,feature_name=tfvocab, categorical_feature= categorial_features)
lgvalid = lgb.Dataset(X_valid,y_valid,feature_name=tfvocab, categorical_feature= categorial_features)
```

```python
lgb_clf = lgb.train(lgb_param,lgtrain, valid_sets=[lgtrain,lgvalid], valid_names=['train','valid'])
```

```python
lgb_clf
```

```python
lgb_clf.predict(X_valid)
```
